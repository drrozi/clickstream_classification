```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Berechne DTW-Pfad f√ºr die Visualisierung mit Python

```python
from dtaidistance import dtw, clustering
from dtaidistance import dtw_visualisation as dtwvis
```

```python
a = np.array([1, 8, 2, 4, 3, 2, 4, 3, 4, 1])
b = np.array([0, 0, 0, 0, 0, 3, 0, 0, 0, 0])
```

```python
path = dtw.warping_path(a,b)
dtwvis.plot_warping(a,b,path)
```

Visualisierung Knotenkopplungen mit gleichschrittigen, Discrete Frechet Distance und DTW. H√§ndisch eingegebene Verbindung nach obigem warping-path

```python
fig, ax = plt.subplots(1,3, figsize=(12,4))

fig.patch.set_visible(False)
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')

xx = [(i, a[i]) for i in np.arange(0, len(a))]
yy = [(j, b[j]) for j in np.arange(0, len(b))]

# gleichschrittige Distanzma√üe
for i, j in zip(xx, yy):   
    ax[0].plot([i[0], j[0]],[i[1], j[1]], '--k', linewidth=1)
    
ax[0].plot(a, '-ro', label='x')#, linewidth=4, markersize=20, markerfacecolor='lightcoral', markeredgecolor='lightcoral')
ax[0].plot(b, '-bo', label='y')#, linewidth=4, markersize=20, markerfacecolor='skyblue', markeredgecolor='skyblue')

## Discrete Frechet Distance
ax[1].plot([1,5], [8,3], '--k', linewidth=1)
ax[1].plot(a, '-ro', label='x')
ax[1].plot(b, '-bo', label='y')

# DTW
ax[2].plot([0,0], [1,0], '--k', linewidth=1)
ax[2].plot([0,1], [1,0], '--k', linewidth=1)
ax[2].plot([0,2], [1,0], '--k', linewidth=1)
ax[2].plot([0,3], [1,0], '--k', linewidth=1)
ax[2].plot([0,4], [1,0], '--k', linewidth=1)

ax[2].plot([1,5], [8,3], '--k', linewidth=1)
ax[2].plot([2,5], [2,3], '--k', linewidth=1)
ax[2].plot([3,5], [4,3], '--k', linewidth=1)
ax[2].plot([4,5], [3,3], '--k', linewidth=1)
ax[2].plot([5,5], [2,3], '--k', linewidth=1)
ax[2].plot([6,5], [4,3], '--k', linewidth=1)
ax[2].plot([7,5], [3,3], '--k', linewidth=1)
ax[2].plot([8,5], [4,3], '--k', linewidth=1)

ax[2].plot([9,6], [1,0], '--k', linewidth=1)
ax[2].plot([9,7], [1,0], '--k', linewidth=1)
ax[2].plot([9,8], [1,0], '--k', linewidth=1)
ax[2].plot([9,9], [1,0], '--k', linewidth=1)

ax[2].plot(a, '-ro', label='x')
ax[2].plot(b, '-bo', label='y')

# fig.savefig("your_dir/dist_vgl.pdf", bbox_inches='tight')
```
